```python
import os
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Update the directory to reflect the correct format
RAW_DATA_DIR = 'archive/raw/'
SAVE_TO_DIR = 'archive/processed/'

# Initialize an empty list to store data from all files
all_data = []
about_raw_data = []

# Iterate over each HTML file in the specified directory
for filename in os.listdir(RAW_DATA_DIR):
    if filename.endswith('.html'):
        filepath = os.path.join(RAW_DATA_DIR, filename)

        # Read the HTML file
        with open(filepath, 'r', encoding='latin-1') as file:
            html_content = file.read()

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Find the table
        table = soup.find('table', id="_ctl0_ContentPlaceHolder1_dgResultado")

        if not table:
            continue

        # Extract table headers
        headers = [th.text for th in table.find_all('tr')[0].find_all('td')]

        # Extract table rows
        rows = []
        for tr in table.find_all('tr')[1:]:
            cells = tr.find_all('td')
            rows.append([cell.text.strip() for cell in cells])


        about_raw_data.append([filename, len(headers), len(rows)])

        # Append rows to all_data
        all_data.extend(rows)

# Create DataFrame from all collected data
if all_data:
    df = pd.DataFrame(all_data, columns=headers)

    # Display the DataFrame
    print(df)

    df.to_csv(SAVE_TO_DIR + 'combined_output.csv', index=False)
else:
    print("No valid tables found in the HTML files.")

if about_raw_data:
    about_raw_data_df = pd.DataFrame(about_raw_data, columns=['filename', 'num_columns', 'num_rows'])
    about_raw_data_df.to_csv(SAVE_TO_DIR + 'about_raw_data.csv', index=False)
```
